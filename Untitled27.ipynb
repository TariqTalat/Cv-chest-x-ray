```python
import os
import shutil
from sklearn.model_selection import train_test_split

source_dir = r"F:\NIH_balanced"
output_dir = r"F:\NIH_split"

classes = ["Edema", "Nodule Mass", "Normal", "Pneumonia"]

for cls in classes:

    cls_path = os.path.join(source_dir, cls)

    images = [f for f in os.listdir(cls_path)
              if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    train_files, temp_files = train_test_split(
        images,
        test_size=0.30,
        random_state=42
    )

    val_files, test_files = train_test_split(
        temp_files,
        test_size=0.50,
        random_state=42
    )

    for split_name, split_files in {
        "train": train_files,
        "val": val_files,
        "test": test_files
    }.items():

        dst_folder = os.path.join(output_dir, split_name, cls)
        os.makedirs(dst_folder, exist_ok=True)

        for file in split_files:
            shutil.copy2(
                os.path.join(cls_path, file),
                os.path.join(dst_folder, file)
            )

print("Split completed successfully!")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

import os

IMG_SIZE = (224,224)
BATCH_SIZE = 16
EPOCHS = 20

train_dir = r"F:\NIH_split\train"
val_dir   = r"F:\NIH_split\val"
test_dir  = r"F:\NIH_split\test"
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=10,
    zoom_range=0.1,
    width_shift_range=0.05,
    height_shift_range=0.05,
    horizontal_flip=True
)

test_datagen = ImageDataGenerator(
    rescale=1./255
)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

val_generator = test_datagen.flow_from_directory(
    val_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

print(train_generator.class_indices)
```

```python
base_model = DenseNet121(
    weights='imagenet',
    include_top=False,
    input_shape=(224,224,3)
)

base_model.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)

x = Dense(512, activation='relu')(x)
x = Dropout(0.5)(x)

output = Dense(4, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=output)
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
callbacks = [

    EarlyStopping(
        monitor='val_loss',
        patience=5,
        restore_best_weights=True
    ),

    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.2,
        patience=2,
        verbose=1
    ),

    ModelCheckpoint(
        "DenseNet121_NIH_best.h5",
        monitor='val_accuracy',
        save_best_only=True,
        verbose=1
    )
]
```

```python
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=EPOCHS,
    callbacks=callbacks
)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator


TRAIN_DIR = r"F:\NIH_split\train"
VAL_DIR = r"F:\NIH_split\val"
TEST_DIR = r"F:\NIH_split\test"
BATCH_SIZE = 32
EPOCHS = 10
IMG_SIZE = (224, 224)
NUM_CLASSES = 4


#  (Data Generators)
# ---------------------------------
#  Data Augmentation 
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    horizontal_flip=True
)

#  (Rescaling)
val_test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    TRAIN_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True
)

val_generator = val_test_datagen.flow_from_directory(
    VAL_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

test_generator = val_test_datagen.flow_from_directory(
    TEST_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)


#  (Feature Fusion Model)

input_shape = (224, 224, 3)
inputs = layers.Input(shape=input_shape)

# النموذج الأول: DenseNet121 لاستخراج الميزات
base_model1 = tf.keras.applications.DenseNet121(weights='imagenet', include_top=False, input_shape=input_shape)
# تجميد الأوزان الأساسية للتعلم المنقول
base_model1.trainable = False 
x1 = base_model1(inputs)
x1 = layers.GlobalAveragePooling2D()(x1)  # تسطيح الميزات

# النموذج الثاني: ResNet50 لاستخراج الميزات
base_model2 = tf.keras.applications.ResNet50(weights='imagenet', include_top=False, input_shape=input_shape)
base_model2.trainable = False
x2 = base_model2(inputs)
x2 = layers.GlobalAveragePooling2D()(x2)  # تسطيح الميزات

# دمج الميزات المستخرجة من الشبكتين معاً أفقياً (Concatenation)
fused_features = layers.concatenate([x1, x2])

# طبقة التصنيف النهائية المشتركة
x = layers.Dense(512, activation='relu')(fused_features)
x = layers.Dropout(0.3)(x)
outputs = layers.Dense(NUM_CLASSES, activation='softmax')(x)

# تجميع النموذج النهائي
model = models.Model(inputs=inputs, outputs=outputs)


#  (Compilation)

model.compile(
    optimizer=optimizers.Adam(learning_rate=1e-4),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

# (Training)

checkpoint = tf.keras.callbacks.ModelCheckpoint(
    'best_tf_fusion_model.keras', 
    monitor='val_accuracy', 
    save_best_only=True, 
    mode='max'
)

history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=val_generator,
    callbacks=[checkpoint]
)

# (Testing)
print("\n=== Testing Final Model ===")
best_model = models.load_model('best_tf_fusion_model.keras')
test_loss, test_acc = best_model.evaluate(test_generator)
print(f"Final Test Accuracy: {test_acc:.4f}")
```
